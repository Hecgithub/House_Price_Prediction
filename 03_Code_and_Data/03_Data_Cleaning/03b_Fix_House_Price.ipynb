### Step 3b (Correct the data price)

The prices of the houses are incorrect due to a mistake in the code of Step 1.
The correct house price is retreived again in this notebook and a new table is to be created

#### Import requested libraries

```python
import pandas as pd
import requests
import json
import re
import os
import glob
from datetime import datetime as dt
import warnings
from time import sleep
from tqdm import tqdm

warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)

# Import your RapidAPI key
from config import krapid as key
```

#### Read the input data

```python
# Load the input data (only the most recent file)
data_path = "data/*.csv"
listing = glob.glob(data_path)
latest_file = max(listing, key=os.path.getctime)
print(f'The file selected is: {latest_file}')
```

```python
# Create the input DF
df_raw = pd.read_csv(latest_file)
```

#### Retreive the data

```python
df_copy = df_raw.copy()

# API specific (do not modify)
host = "zillow-com1.p.rapidapi.com"
url = "https://" + host + "/property" # Query of single property details

# Create an empty DF, only declare colun names
cnames = ['zid', 'priceOK']
df = pd.DataFrame(columns=cnames)

for i in df_copy.index:
    zid = df_copy.at[i, 'zid']
    price = df_copy.at[i, 'Price']    
    
    # API specific (do not modify)
    querystring = {"zpid":zid}

    headers = {
        "X-RapidAPI-Key": key,
        "X-RapidAPI-Host": host
    }
    
    # Max 2 calls per second are allowed
    sleep(0.6)

    response = requests.request("GET", url, headers=headers, params=querystring)
    json_response = response.json()

    # If the response is successful (i.e., 200), then get the property details
    if response.status_code == 200:
        
        # Get the correct house price
        priceOK = json_response['price']
        
        # Append to output DF
        df_row = {'zid': zid, 'priceOK': priceOK}
        df = df.append(df_row, ignore_index=True)

    else:

        priceOK = price
        
        # Append to output DF
        df_row = {'zid': zid, 'priceOK': priceOK}
        df = df.append(df_row, ignore_index=True)
              
    print(f'working in item {i} of 4501, zid: {zid}, old price: {price}, Correct price: {priceOK}')
```

#### Merge the DFs and save the data

```python
# Add the correct price to the original DF
df_fix = df_raw.merge(df, on='zid', how='left')

# Output file name
name = latest_file.split("data")[2]
out_name = 'fix_price_cleaned_data'+ name

# Create output directory if it does not exist
os.makedirs('./data/', exist_ok=True)

# Save the merged DF to a csv file
df_fix.to_csv(f'./data/{out_name}', index=False)
```
